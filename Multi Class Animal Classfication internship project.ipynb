```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("iamsouravbanerjee/animal-image-dataset-90-different-animals")

print("Path to dataset files:", path)
```

```python
dataset_path = "/kaggle/input/animal-image-dataset-90-different-animals/animals/animals"
classes = os.listdir(dataset_path)
num_classes = len(classes)

# Display the number of classes along with names
print(f'Number of Classes: {num_classes}')
print(f'Classes: {classes}')
```

```python

# Visualize some sample images form the dataset
plt.figure(figsize=(15,10))
for i, class_name in enumerate(classes[:10]):
  class_path = os.path.join(dataset_path, class_name)
  img_name = os.listdir(class_path)[0]
  img_path = os.path.join(class_path, img_name)
  img = plt.imread(img_path)

  plt.subplot(2, 5, i+1)
  plt.imshow(img)
  plt.title(f'{class_name} \n shape: {img.shape}')
  plt.axis('off')
plt.show()
```

```python
class_name = input("Enter species name: ")
dataset_path = "/kaggle/input/animal-image-dataset-90-different-animals/animals/animals"
selected_class = class_name  # Use the user's input
class_path = os.path.join(dataset_path, selected_class)

# Get all image file names from the selected class
image_files = os.listdir(class_path)

# Display some images from the selected class
plt.figure(figsize=(12, 6))

num_images = min(10, len(image_files))  # Display up to 10 images or the max available
for i in range(num_images):
    img_path = os.path.join(class_path, image_files[i])
    img = plt.imread(img_path)

    plt.subplot(2, 5, i + 1)  # Arrange images in a grid
    plt.imshow(img)
    plt.title(f'Various {selected_class}   \n shape: {img.shape}')
    plt.axis('off')

plt.show()
```
